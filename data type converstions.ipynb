# Data Type Converstions

```python
#data type converstions

#integers

a=1
b=2
c=3
d=4
num1=5
num2=6
num3=7
num4=8
print("integers")
print(a)
print(b)
print(c)
print(d)
print(num1)
print(num2)
print(num3)
print(num4)

#float

AB=1.00
BC=2.00
CD=3.11
DE=4.45
EF=5.65
GH=6.78
IJ=7.45
KL=8.78
MN=9.34
print("float")
print(AB)
print(BC)
print(CD)
print(DE)
print(EF)
print(GH)
print(IJ)
print(KL)
print(MN)

#complex

AB = 24+0j
BC = 46+1j
CD = 78+20j
DE = 56+25j
FG = 43+32j
HI = 245+3456j
GH = 23+23j
JK = 23+21j
LM = 46+45j
DF = 222+23j
print("complex")
print(AB)
print(BC)
print(CD)
print(DE)
print(FG)
print(HI)
print(GH)
print(JK)
print(LM)
print(DF)

#type converstions
#int to float

a=float(10)
b=float(20)
c=float(30)
d=float(40)
num1=float(50)
num2=float(60)
num3=float(70)
num4=float(80)
print("--int to float--")
print(a)
print(b)
print(c)
print(d)
print(num1)
print(num2)
print(num3)
print(num4)

#float to int

AB=int(1.00)
BC=int(2.00)
CD=int(3.11)
DE=int(4.45)
EF=int(5.65)
GH=int(6.78)
IJ=int(7.45)
KL=int(8.78)
MN=int(9.34)
print("---float to int---")
print(AB)
print(BC)
print(CD)
print(DE)
print(EF)
print(GH)
print(IJ)
print(KL)
print(MN)

#int to complex

a=complex(10)
b=complex(20)
c=complex(30)
d=complex(40)
num1=complex(50)
num2=complex(60)
num3=complex(70)
num4=complex(80)
print("---int to complex---")
print(a)
print(b)
print(c)
print(d)
print(num1)
print(num2)
print(num3)
print(num4)

#float to complex

AB=complex(1.00)
BC=complex(2.00)
CD=complex(3.11)
DE=complex(4.45)
EF=complex(5.65)
GH=complex(6.78)
IJ=complex(7.45)
KL=complex(8.78)
MN=complex(9.34)
print("---float to complex---")
print(AB)
print(BC)
print(CD)
print(DE)
print(EF)
print(GH)
print(IJ)
print(KL)
print(MN)




```
