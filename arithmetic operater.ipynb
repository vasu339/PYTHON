```python
# Define 10 numbers
a = 4
b = 9
c = 20
d = 10
e = 7
f = 5
g = 12
h = 3
i = 8
j = 15
# Addition
add1 = a + b
add2 = c + d
add3 = e + f
add4 = g + h
add5 = i + j
print("Addition results:")
print(add1)
print(add2)
print(add3)
print(add4)
print(add5)
# Subtraction
sub1 = a - b
sub2 = c - d
sub3 = e - f
sub4 = g - h
sub5 = i - j
print("\nSubtraction results:")
print(sub1)
print(sub2)
print(sub3)
print(sub4)
print(sub5)
# Multiplication
mul1 = a * b
mul2 = c * d
mul3 = e * f
mul4 = g * h
mul5 = i * j
print("\nMultiplication results:")
print(mul1)
print(mul2)
print(mul3)
print(mul4)
print(mul5)
# Division
div1 = a / b
div2 = c / d
div3 = e / f
div4 = g / h
div5 = i / j
print("\nDivision results:")
print(div1)
print(div2)
print(div3)
print(div4)
print(div5)
# Modulus
mod1 = a % b
mod2 = c % d
mod3 = e % f
mod4 = g % h
mod5 = i % j

print("\nModulus results:")
print(mod1)
print(mod2)
print(mod3)
print(mod4)
print(mod5)

# Exponentiation
exp1 = a ** b
exp2 = c ** d
exp3 = e ** f
exp4 = g ** h
exp5 = i ** j

print("\nExponentiation results:")
print(exp1)
print(exp2)
print(exp3)
print(exp4)
print(exp5)

# Floor Division
floor_div1 = a // b
floor_div2 = c // d
floor_div3 = e // f
floor_div4 = g // h
floor_div5 = i // j

print("\nFloor Division results:")
print(floor_div1)
print(floor_div2)
print(floor_div3)
print(floor_div4)
print(floor_div5)
```
